# 3.1 Variables & Assignments - You learned how to store information in variables, like putting a label on a box to remember what's inside.

# 3.4 Strings - You learned how to work with text (words and sentences) in your code.
# 3.5 Boolean Expressions - You learned how to create true/false statements to help your program make decisions.
# 3.6 Conditionals - You learned how to make your program do different things based on conditions, like "if this is true, then do that."
# 3.7 Nested Conditionals - You learned how to put if-statements inside other if-statements for more complex decisions.
# 3.8 Iteration - You learned how to repeat actions multiple times using loops instead of writing the same code over and over.
# 3.9 Developing Algorithms - You learned how to create step-by-step instructions to solve problems.
# 3.10 Lists - You learned how to store multiple items in one variable, like a shopping list.
# 3.12 Calling Procedures - You learned how to use pre-made functions that someone else created to do tasks quickly.
# 3.13 Developing Procedures - You learned how to create your own reusable functions so you don't repeat code.
# 3.15 Random Values - You learned how to generate random numbers for things like games or simulations.
# 3.17 Algorithmic Efficiency - You learned how to write code that runs faster and uses less computer resources.
