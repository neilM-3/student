## Summary
Reworded interview notes capturing the interviewee’s motivation for their major, how high school prepared them, teamwork/process habits, and advice for students entering CS.

---

## Major Inspiration / Why They Chose the Major
- Grew up in/near the Bay Area, so exposure to tech played a role early on.
- Mentioned Legos as an early influence: enjoying creativity and building things from scratch.

---

## High School Preparation for College-Level Work
- Stressed that communication is essential.
- Writing documentation/reports helped them practice clearly explaining technical work.
- Participated in CyberPatriot, which helped them become more comfortable with computers.

---

## Breaking Down Big Problems
- Said decomposing large tasks into smaller steps is a major skill in their field.
- Example: splitting essays into steps to make work manageable.
- Example: dividing projects by components (e.g., frontend vs backend) and assigning owners to reduce workload and stay organized.

---

## Retrospectives / Project Reviews
- Has done retrospectives/reviews and found them valuable for identifying what worked, what didn’t, and what to change next time.

---

## Challenging Technical/Engineering Problem + Approach
- Highlighted the difficulty of managing large projects/heavy workloads.
- Approach: break the project into clear parts, delegate responsibilities, and coordinate regularly so the pieces still fit together.

---

## Project Ideas + Creativity vs Constraints
- Uses user research to generate ideas and guide decisions.
- Emphasizes a “users first” mindset.
- Noted internships often provide fewer creative choices since tasks are assigned.
- In class/personal projects, they have more freedom—still improved work by observing user behavior and iterating based on feedback.

---

## Strongest Part of Their College Application
- Focused on growth and what they learned in high school.
- Connected their experiences to why they were a good match for the colleges they applied to.
- Included personal story elements and impact on others (e.g., starting a nonprofit / working with other organizations).
- Framed CS interest around exploration, learning, and helping others through what they build.

---

## Most Valuable Skills from High School + What They Wish They Learned Earlier
- STEM classes built lasting skills: time management, teamwork, communication, and mindset.
- Emphasized grit and the ability to work hard.
- Said a rigorous school environment helped them handle later challenges.
- Wishes they learned earlier how to take initiative with the freedom given in projects.
- Mentioned learning to use resources effectively (Stack Overflow, ChatGPT, etc.) and being proactive about solving problems.

---

## Communication in Projects (Talk-Throughs, Diagrams, Documentation)
- Talking through code helps align the team and clarify ideas.
- AI tools can support understanding or debugging code.
- Diagrams/flowcharts help visualize systems and improve clarity.
- Sharing written docs about new concepts helps the rest of the team ramp up faster.
- Keeping user needs in mind supports clearer communication and better decisions.

---

## Agile/Scrum + Teamwork
- Has used agile/scrum-like methods and found they help structure work.
- These methods support organizing people and breaking tasks into manageable chunks.
- Reported both challenges and successes in team settings; earlier group projects helped prepare them for college teamwork.

---

## Advice Shared
- Stay updated on new CS developments since the field changes quickly and spans many areas.
- Lower stress and staying relaxed can improve learning and understanding.
- “If you really want something, you can achieve it.”
- Try new things while you still can and say yes to opportunities.
- Take advantage of project-based classes and the freedom they provide.
