## Extra Credit: Proble Book List Manipulation

```python
%%js
import java.util.ArrayList;

public class BookListManipulation {
    public static void main(String[] args) {
        // A list (ArrayList) of books
        ArrayList<String> bookList = new ArrayList<>();
        bookList.add("1984");
        bookList.add("Brave New World");
        bookList.add("The Great Gatsby");
        bookList.add("Moby Dick");
        bookList.add("To Kill a Mockingbird");

        // 1. Print the list of books
        System.out.println("Books: " + String.join(", ", bookList));

        // 2. Add a new book
        bookList.add("Pride and Prejudice");
        System.out.println("After adding 'Pride and Prejudice': " + String.join(", ", bookList));

        // 3. Remove a book
        bookList.remove("Moby Dick");
        System.out.println("After removing 'Moby Dick': " + String.join(", ", bookList));

        // 4. Print each book name
        System.out.println("\nUpdated Book List:");
        for (String book : bookList) {
            System.out.println(book);
        }
    }
}
```
